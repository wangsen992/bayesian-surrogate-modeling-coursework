## Problem 3

```python
N1 = 20
N0 = 10
a1 = 20
b1 = 20
a2 = 10
b2 = 10
```

```python
theta = np.arange(0,1,0.01)
```

```python
likelihood = comb(N1+N0,N1) * theta**(N1) * (1-theta)**N0*40
prior1 = beta.pdf(theta, a1, b1)
prior2 = beta.pdf(theta, a2, b2)
evidence = comb(N1+N0, N1)*(beta_B(a1,b1)+beta_B(a2,b2))*beta_B(N1+a1,N0+b1)*beta_B(N1+a2,N0+b2)/\
           (beta_B(a1,b1)*beta_B(a2,b2)*(beta_B(N1+a1,N0+b1) + beta_B(N1+a2, N0+b2)))
posterior = (0.5*beta.pdf(theta, N1+a1, N0+b1) + 0.5*beta.pdf(theta, N1+a2, N0+b2))
```

```python
fig, ax = plt.subplots(figsize=(10,8))
p1, = ax.plot(theta, prior1, label='Beta(20,20)')
p2, = ax.plot(theta, prior2, label='Beta(10,10)')
li, = ax.plot(theta, likelihood, '--',label='likelihood')
pos, = ax.plot(theta, posterior, label='Posterior')
ax.legend(handles=[p1,p2,pos, li])
plt.savefig("prob3b.png")
plt.show()
```

```python
def phi(x, M, s=0):
    "Return features based on polynomial degree M from single input x"
    return np.array([[x**i for i in range(s,M+1)]])

def Phi(X, M, s=0):
    return np.concatenate([phi(x,M,s) for x in X], axis=0)
```

```python
N = np.asscalar(mat7['N'])
x = mat7['x'][:,0]
y = mat7['y']
a = 0.1
b = 0.00001
gam = 0.001/N
```

### Part B
For polynomial orders $M = 1,2,3,4,5$ plot the predictive mean and the predictive error bars. Your plots should also indicate the exact functoin as well as the training data points.

```python
import ipdb as debugger
```

```python
class GigModel(object):
    
    def __init__(self, X_in, Y_in, M, centered=False):
        self.X_in = X_in
        self.Y_in = Y_in
        self.M = M
        self.N = X_in.shape[0]
        
        if centered:
            self.__train_centered()
        else:
            self.__train()
        self.__evidence()
    
    def __train(self):
        # Calculate predictive distribution parameters
        Phi_in = Phi(self.X_in, self.M)
        V_N_inv = 1/gam * np.eye(Phi_in.shape[1]) + Phi_in.T @ Phi_in
        self.V_N = np.linalg.inv(V_N_inv)
        print(f"Condition number for V_N_inv is {np.linalg.cond(V_N_inv):f}")
        #debugger.set_trace()
        self.w_N = self.V_N @ Phi_in.T @ self.Y_in
        self.a_N = a + self.X_in.shape[0]/2
        self.b_N = np.asscalar(b + 0.5 * (self.Y_in.T @ self.Y_in - self.w_N.T @ V_N_inv @ self.w_N))
        print(f"a_N = {self.a_N:f}")
        print(f"b_N = {self.b_N:f}")   
    
    def __train_centered(self):
        # Center input data
        Phi_in = Phi(self.X_in, self.M, s=1)
        Phi_in_mean = np.mean(Phi_in, axis=0)
        Phi_in_c = Phi_in - Phi_in_mean
        Y_in_mean = np.mean(self.Y_in, axis=0)
        Y_in_c = self.Y_in - Y_in_mean
        
        # Calculate predictive distribution parameters
        V_N_inv = 1/gam * np.eye(Phi_in_c.shape[1]) + Phi_in_c.T @ Phi_in_c
        V_N = np.linalg.inv(V_N_inv)
        print(f"Condition number for V_N_inv is {np.linalg.cond(V_N_inv):f}")
        #debugger.set_trace()
        w_N_c = V_N @ Phi_in_c.T @ Y_in_c
        self.a_N = a + self.X_in.shape[0]/2
        self.b_N = np.asscalar(b + 0.5 * (Y_in_c.T @ Y_in_c - w_N_c.T @ V_N_inv @ w_N_c))
        w0 = Y_in_mean - Phi_in_mean @ w_N_c
        self.w_N = np.insert(w_N_c,0,w0).reshape(self.M+1,1)
        self.V_N = np.zeros([self.M+1, self.M+1])
        self.V_N[1:,1:] = V_N
        print(f"a_N = {self.a_N:f}")
        print(f"b_N = {self.b_N:f}")  
        
        
    def predict(self, X_out):
        # Calculate output
        Phi_out = Phi(X_out, self.M)
        mean_out = (Phi_out @ self.w_N)[:,0]
        df = 2*self.a_N
        std_out = np.diagonal(self.b_N / self.a_N * (np.eye(X_out.shape[0]) + Phi_out @ self.V_N @ Phi_out.T))        

        return (mean_out, np.sqrt(std_out))
    
    def sample(self, X_out, N, ax=None):
        "Produce N sample results from the model"
        V = self.b_N / self.a_N * np.diagonal(self.V_N)
        W = np.array([t.rvs(df=2*self.a_N, loc=w_i, scale=v1, size=N) for w_i, v1 in zip(self.w_N[:,0], V)])
        Phi_out = Phi(X_out, self.M)
        Y_out = Phi_out @ W
        if ax is None:
            return Y_out
        else:
            for i in range(N):
                ax.plot(X_out, Y_out[:,i],'r')
            ax.plot(self.X_in, self.Y_in, 'o')
            ax.set_title(f"M={self.M}")
    
    def __evidence(self):
        self.evidence = (2*np.pi)**(-self.N/2) * gam**(-self.M/2) * np.sqrt(np.linalg.det(self.V_N))\
                        * gamma(self.a_N)/gamma(a) * b**a / (self.b_N**self.a_N)
        
        
        
        
```
